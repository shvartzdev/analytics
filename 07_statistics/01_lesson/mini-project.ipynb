```python
gb_df = df.loc[df['Publisher'] == 'Nintendo', ['Year', 'Genre','Global_Sales']]
```

```python
genres = ['Sports']

gb_df = gb_df.query('Genre in @genres').groupby(by='Year', as_index=False).agg({'Global_Sales':'sum'})
```

```python
gb_df
```

```python
sns.set(rc={"figure.figsize":(22, 16)}) #width=3, #height=4
sns.lineplot(data=gb_df, x="Year", y="Global_Sales")
```
