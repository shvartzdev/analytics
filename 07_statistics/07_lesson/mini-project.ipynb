### Задание 6

Теперь построим ещё две модели:

- модель со всеми предикторами  
- модель со всеми предикторами, кроме марок машин  

Попробуем выбрать из этих моделей самую удачную! В каком-то смысле это искусство - при выборе моделей приходится ориентироваться не только на количественные показатели, но также на контекст её использования, опыт и здравый смысл.

````python
predictors = df_dummy_full.columns[df_dummy_full.columns != 'price'].values
````

````python
X = sm.add_constant(df_dummy_full[predictors])
````

````python
Y = cars.price
````

````python
model = sm.OLS(Y, X).fit()
````

````python
print(model.summary())
````

````python
summary = model.summary()

coef_table = summary.tables[1].as_html()
df_coef = pd.read_html(coef_table, header=0, index_col=0)[0]
````

````python
df_coef.shape
````

````python
# Фильтрация статистически значимых переменных (P>|t| < 0.05)
significant_variables = df_coef[df_coef['P>|t|'] < 0.05]
print(significant_variables)
````

````python
len(significant_variables)
````

````python
unsignificant_variables = df_coef[df_coef['P>|t|'] > 0.05]
len(unsignificant_variables)
````
